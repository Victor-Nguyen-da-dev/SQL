```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast
from adjustText import adjust_text

df = pd.read_csv('Query 1.csv')
df
```

```python
sns.barplot(data=df, x='salary_year_avg', y='job_title', hue='salary_year_avg', palette='Blues', legend=False, edgecolor='black')
sns.set_theme(style='ticks')
sns.despine()
plt.xlabel('Average Yearly Salary ($USD)')
plt.ylabel('')
plt.title('Average Salary Distribution of the Top 10 Paying Data Analyst Jobs in 2023')
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))

for y, x in enumerate(df['salary_year_avg']):
    plt.text(x + 15000, y, f'${int(x/1000)}K', va='center')

plt.show()
```
